```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "C://Users//Logan//Desktop//bootcamp//API_Challenge//python-api-challenge//output_data//"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Starting api Call
api_URL = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
#blank list for cities and starting counts at 1
api_Cities_List = []
api_Record_count = 1
api_set_count = 1
# Start print log with this 
print("**************************")
print(" Starting Record Retrieval")
print("**************************")
#for loop through all cities
for city in cities:
      #add city to city url
    city_URL = api_URL + "&q=" + city
    #run an api request for each of the cities
    try:
        #getting city json response
        city_json = requests.get(city_URL).json()
        #increasing record count 
        api_Record_count +=1
        #get the temp, humidity, clouds, wind speed, country, date and latitude/longitude
        city_Latitude  = city_json["coord"]["lat"]
        city_longitude = city_json["coord"]["lon"]
        city_Tempature = city_json["main"]["temp"]
        city_Humidity = city_json["main"]["humidity"]
        city_Clouds = city_json["clouds"]["all"]
        city_Wind = city_json["wind"]["speed"]
        city_country = city_json["sys"]["country"]
        city_date = city_json["dt"]
        # append city info to the api_Cities_List data list
        api_Cities_List.append({"City": city,
                                "Latitude" : city_Latitude,
                                "Longitude" : city_longitude,
                                "Temperature" : city_Tempature,
                                "Humidity" : city_Humidity,
                                "Cloudiness" : city_Clouds,
                                "Wind Speed" : city_Wind,
                                "Country" : city_country,
                                "Date": city_date})
        print(f"Processing Record {api_Record_count} of set {api_set_count} | {city}")
        #If error increase count and print skip
    except:
            api_Record_count = 0
            api_set_count +=1
            print("City is not Avalible, Skip to next city")           
            pass
#print end of log
print("**************************")        
print("Ending Record Retrieval")
print("**************************")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#converting to Dataframe
api_Cities_List_df = pd.DataFrame(api_Cities_List)
#export in to csv
api_Cities_List_df.to_csv(output_data_file + "cities.csv")

api_Cities_List_df.head()
```

## Linear Regression

```python
#creating hemisphere
northern_Hemisphere = api_Cities_List_df2.loc[api_Cities_List_df2["Latitude"] > 0.01]
southern_Hemisphere = api_Cities_List_df2.loc[api_Cities_List_df2["Latitude"] < 0.01]
#northern_Hemisphere.head()
#southern_Hemisphere.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#setting x and y values 
x_values = northern_Hemisphere["Latitude"]
y_values = northern_Hemisphere["Temperature"]
#Linregress formula
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
#ploting scatter with regression visible and equataion
plt.scatter(x_values,y_values)
plt.plot(x_values, regress_values, "orange")
plt.annotate(line_eq,(0,0), fontsize = 12, color="orange")
#creating title
plt.title("Temperature vs. Latitude "+ time.strftime("%x") )
#lables
plt.ylabel("Temperature")
plt.xlabel("Latitude")
#Saving figure
plt.savefig(output_data_file + "northerntempRegression.png")
```
